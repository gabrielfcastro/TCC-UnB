Categorização em ataques

1. Categoria: MQTT Attack (IOT)

Ataques que exploram o protocolo MQTT, amplamente usado em dispositivos IoT. Incluem floods de conexão/publicação e injeção de pacotes malformados, visando exaustão de recursos, indisponibilidade do serviço e possíveis falhas de segurança em brokers e dispositivos conectados.

2. Categoria: Reconnaissance (Scanning)

Atividades de coleta de informações sobre a rede ou dispositivos, como varredura de portas, sistemas operacionais e vulnerabilidades. Geralmente são etapas iniciais de ataques, usadas para mapear alvos antes da exploração efetiva.

```python
# Pré-processamento direto no DF
df_atack_category.columns = df_atack_category.columns.str.replace(' ', '_', regex = False)
df_content = df_atack_category.select_dtypes(include = np.number).columns
df_atack_category[df_content] = df_atack_category[df_content].replace([np.inf, -np.inf], np.finfo(np.float32).max)
df_atack_category[df_content] = df_atack_category[df_content].fillna(0)
df_atack_category.columns

# Garante índice único
df_atack_category = df_atack_category.reset_index(drop = True)

def balanced_df(df, n_per_class, file_name):
  index_to_remove = []
  df_list = []

  for label in df['Label'].unique():
    group = df[df['Label'] == label]
    n_safe = min(len(group), n_per_class)

    if n_safe < n_per_class:
      print(f"Aviso: Classe {label} só tem {n_safe} linhas restantes.")

    sample = group.sample(n = n_safe, random_state = 42)
    df_list.append(sample)
    index_to_remove.extend(sample.index)

  df_final = pd.concat(df_list, ignore_index = True)
  df_final = shuffle(df_final, random_state = 42)
  df_final.to_csv(file_name, index = False)
  files.download(file_name)

  df = df.drop(index_to_remove)

  return df

def unbalanced(df, n, file_name):
  sample = df.sample(n = n, random_state = 42)
  sample.to_csv(file_name, index = False)
  files.download(file_name)
  df = df.drop(sample.index)
  return df
```
